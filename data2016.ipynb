```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress

csvdata2016= "2016 Stack Overflow Survey Responses.csv"
data2016 = pd.read_csv(csvdata2016)
```

```python
datafor2016 = data2016[['country','age_range','salary_range','tech_do']]
datafor2016=datafor2016.dropna()
```

```python
clean2016 = datafor2016.reset_index()
```

```python
clean2016 = clean2016[['country', 'age_range', 'salary_range', 'tech_do']]
clean2016
```

```python
jezik = clean2016[['tech_do']]

jezik['tech_do'] = jezik['tech_do'].str.replace(' ', '')
jezik
```

```python
def split_multicolumn(col_series):
    result_df = col_series.to_frame()
    options = []
    # Iterate over the column
    for idx, value  in col_series[col_series.notnull()].iteritems():
        # Break each value into list of options
        for option in value.split(';'):
            # Add the option as a column to result
            if not option in result_df.columns:
                options.append(option)
                result_df[option] = False
            # Mark the value in the option column as True
            result_df.at[idx, option] = True
    return result_df[options]
```

```python
languages2 = split_multicolumn(jezik.tech_do)

languages2
```

```python
temp = clean2016[['country', 'age_range', 'salary_range']]
clean_2016 = temp.join(languages2, how='outer')

clean_2016
```

```python
languages_worked_percentages = languages2.mean().sort_values(ascending=False) * 100
language_percent = languages_worked_percentages.reset_index()
language_percent.rename({"index": "language", 0: "percentage"}, axis=1, inplace=True)
language_percent_2016 = language_percent
language_percent_2016
```

```python
language_percent_2016.to_csv("language_percent_2016.csv", index=True, header=True)
```

```python
clean_2016.to_csv("clean_2016.csv", index=True, header=True)
```
