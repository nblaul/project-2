```python
lang_one = "clean_data/2021_avg_salary_by_language.csv"
lang_one_df = pd.read_csv(lang_one, encoding="ISO-8859-1")
lang_one_df.shape

```

```python
lang_two = "clean_data/2020_avg_salary_by_language.csv"
lang_two_df = pd.read_csv(lang_two, encoding="ISO-8859-1")
lang_two_df.shape
```

```python
lang_three = "clean_data/2019_avg_salary_by_language.csv"
lang_three_df = pd.read_csv(lang_three, encoding="ISO-8859-1")
lang_three_df.shape
```

```python
lang_four = "clean_data/2018_avg_salary_by_language.csv"
lang_four_df = pd.read_csv(lang_four, encoding="ISO-8859-1")
lang_four_df.shape
```

```python
lang_five = "clean_data/2017_avg_salary_by_language.csv"
lang_five_df = pd.read_csv(lang_five, encoding="ISO-8859-1")
lang_five_df.shape
```

```python
frames = [lang_one_df, lang_two_df, lang_three_df, lang_four_df, lang_five_df]

lang_df = pd.concat(frames)
lang_df
```

```python
lang_df.columns
```

```python
lang_df.drop('Unnamed: 0', axis=1, inplace=True)
lang_df = lang_df.reset_index()
lang_df
```

```python
# use this to poppulate mongo 'salary' collection
lang_df.to_csv("clean_data/lang_df.csv", index=False, header=True)
```
